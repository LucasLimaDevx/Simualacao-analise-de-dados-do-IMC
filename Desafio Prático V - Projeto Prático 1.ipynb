### Importando pacotes

```python
import pandas as pd
import numpy as np
from sympy import *
```

### Definindo tamanho da amostra

```python
N = 3000
```

### Gerando dados das alturas em metros

```python
#Gerando alturas com média de 1.70 e desvio de 0.25 em uma distribuição normal, para 3000 pessoas
array_alturas = np.random.normal(1.70, 0.27, N)
array_alturas
```

```python
#Convertendo array de alturas uma lista
lista_alturas = array_alturas.tolist()
```

```python
#Visualizando os primeiros 20 valores da lista de alturas
print(lista_alturas[0:20])
```

### Gerando os pesos com base nos valores das alturas

```python
#Escrevendo valor a função para gerar os valores dos pesos associados para cada altura
altura_i = symbols("altura_i")
valor_aleatorio_i = symbols("valorAleatorio_i")
peso = altura_i * 54 + valor_aleatorio_i
peso
```

```python
from random import *
lista_pesos = []
for altura in lista_alturas:
    valor_aleatorio = round(float(uniform(-6, 6)),2)
    peso = altura * 54 + valor_aleatorio
    lista_pesos.append(peso)
```

```python
# Verificando se foi gerando os pesos para as 3000 amostras
len(lista_pesos)
```

### Escrevendo a fórmula do IMC

```python
peso, altura = symbols(["peso", "altura"])
IMC = peso / altura**2
IMC
```

```python
lista_IMC = []
for peso, altura in zip(lista_pesos, lista_alturas):
    IMC = round(peso / altura**2, 2)
    lista_IMC.append(IMC)
```

```python
lista_IMC
```

```python
# Verificando total IMC gerado
len(lista_IMC)
```

### Gerando sexo aleatoriamente

```python
lista_sexo = []
for i in range(0, 3000):
    valor = np.random.uniform(0, 1)
    if valor > 0.5:
        lista_sexo.append('M')
    else:
        lista_sexo.append('F')
```

```python
lista_sexo
```

```python
# Verificando total de valores relacionado ao sexo que foram gerados
len(lista_sexo)
```

### Gerando lista de pessoas com problemas cardíacos de acordo com carda valor de IMC

```python
lista_cardiacos = []
for i in range(0, 3000):
    if lista_IMC[i] > 40:
        lista_cardiacos.append('S')
    else:
        if lista_IMC[i] >= 30 and lista_IMC[i] <= 40:
            lista_cardiacos.append('P')
        else:
            lista_cardiacos.append('N')
```

```python
lista_cardiacos
```

```python
# Verificando total de valores da lista de cardíacos que foram gerados
len(lista_cardiacos)
```

### Gerando a tabela final

```python
tabela = lista_alturas, lista_pesos, lista_IMC, lista_sexo, lista_cardiacos
len(tabela)
```

```python
df = pd.DataFrame(tabela, index=['Altura', 'Peso', 'IMC', 'Sexo', 'Cardíaco']).T
```

```python
df
```

```python
df.to_csv('df_desafio_prático_multidisciplinar_1.csv')
```

## Determinando a média de tendecias

```python
def media(coluna):
    soma = sum(coluna)
    media = soma / N
    media = round(media, 2)
    return media
```

```python
# Calculando média da Altura
print(media(df.Altura))
```

```python
# Calculando média Peso
print(media(df.Peso))
```

```python
# Calculando média IMC
print(media(df.IMC))
```

### Calculando frênqucia para valores qualitativos

```python
# Calculando frequências de sexos
frequencia_masculino = 0;
frequencia_feminino = 0;

for sexo in lista_sexo:
    if sexo == 'M':
        frequencia_masculino += 1
    else:
        frequencia_feminino += 1

percentual_masculino = (frequencia_masculino / N) * 100
percentual_feminino = (frequencia_feminino / N) * 100

print(f"Percentual de homens é {round(percentual_masculino, 2)}%")
print(f"Percentual de mulheres é {round(percentual_feminino, 2)}%")
```

```python
# Calculando frequência de pessoas problemas cardíacos
frequencia_s = 0
frequencia_p = 0
frequencia_n = 0

for cardiaco in lista_cardiacos:
    if cardiaco == 'S':
        frequencia_s += 1
    elif cardiaco == 'P':
        frequencia_p += 1
    else:
        frequencia_n += 1

percentual_s = (frequencia_s/ N) * 100
percentual_p = (frequencia_p/ N) * 100
percentual_n = (frequencia_n/ N) * 100

print(f"Percentual de pessoal com problemas cardíacos {round(percentual_s, 2)}%")
print(f"Percentual de pessoal com tendencias a problemas cardíacos {round(percentual_p, 2)}%")
print(f"Percentual de pessoal sem problemas cardíacos {round(percentual_n, 2)}%")
```
